# 데이터 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["font.family"] = 'MalgunGothic'
plt.rcParams["font.size"] = 10
```

### 1. 라인 플롯

```python
plt.plot([1, 4, 9, 16]) # X=[0, 1, 2, 3]을 가정, plot([]) 안의 값은 y값
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])  #([],[]) 앞의 []는 x값, 뒤의 []는 y값
plt.title('y = xxxx')
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16], 'rs--')  #([],[]) 앞의 []는 x값, 뒤의 []는 y값
plt.title("'rs--' 스타일")
plt.grid()
plt.show()
```
